```
def k_nearest_neighbours(k, dataset):
  print(" Please enter new x and new y coordinates ")
  new_point = np.array(( int(input()) , int(input()) ))
  dist = []
  for i in dataset:
    dist.append((np.linalg.norm(new_point-i), i))
  dist.sort()
  k_points = []
  for x,y in dist:
    if(k>0):
      k_points.append(y)
      k=k-1
  k_points = np.array(k_points)
  return k_points
```

```
#Getting k value as user input
k=int(input())
answer = k_nearest_neighbours(k, x)
print(answer)
```

#C) Find all points that make positive angle with the new x and y coordinates


```
def pos_angle_points(points):

  print("plotting the 100 points")
  plt.scatter(points[:,0], points[:, 1])

  print(" x new and y new coordinates")
  #new_x = float(input())
  #new_y = 

  p = np.array(( float(input()) , float(input()) ))
  pos_angle = []
  for a in points:
    if np.dot(p, a)>0:
      pos_angle.append(a)

  pos_angle = np.array(pos_angle)
  
  return pos_angle, p
```

```
#generating 100 points
points = gen_num(100)
#print(points)
pos_points, p = pos_angle_points(points)
```

```
running_avg = []
sum = 0
for j in range(100):
  sum+= y[j]
  running_avg.append(sum/(j+1))
running_avg
```

```
#plt.plot(running_avg, y)
x_axis = np.linspace(1, 100, 100)
f1 = []
f2 = []
for i in range(100):
  f1.append( math.sqrt(1/x_axis[i]) )
  f2.append( -1 * math.sqrt(1/x_axis[i]) )
plt.plot(running_avg)
plt.plot(x_axis, f1, color = 'cyan')
plt.plot(x_axis, f2, color = 'magenta')
```
