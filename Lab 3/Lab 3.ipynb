storing the contents of the file into a variable called lines in lower case

```
lines = lines.lower()
lines
```

Splitting the lines in the textfile into singular words

```
words =lines.split()
```

```
words
```

```
import string
```

Table for mapping punctuations to null character

```
table = str.maketrans('','', string.punctuation)
table
```

```
mod_words = [w.translate(table) for w in words]
mod_words
```

```
final_words = [x.strip('') for x in mod_words ]
final_words
```

Getting unique list of words


```
unique_words = set(final_words)
```

```
unique_words
```

```
type(final_words)
```

only_words list contains all the words without emojis and punctuations
dictionary list contains all the unique words

```
only_words = []
dictionary = [] 
for word in final_words:
  letters = list([c for c in word if c.isalpha()])
  word = ''.join(letters)
  if word != '':
    only_words.append(word)
  if word not in dictionary and word!= '':
    dictionary.append(word)
```

```
only_words
```

```
dictionary
```

Number of unique words

```
len(dictionary)
```

Total number of words

```
len(only_words)
```

Creating a dictionary called m_dict which stores the freq of each word

```
my_dict = {}
for unq_word in dictionary:
  count=0
  for word in only_words:
    if unq_word==word:
      count+=1
  my_dict[unq_word]=count
```

```
len(my_dict)
```

#Q2) Making an N X N matrix

```
#importing libraries
import numpy as np
import pandas as pd
```

2D matrix which will store the freq of occurence of the words in column based on the corresponding word in the row 

```
Bigram_matix = np.zeros((len(my_dict),len(my_dict)))
```

```
Bigram_matix
```

```
Bigram_matix.shape
```

A dictionary to store indices for our unique words

```
indices = {}
x=0
for key in my_dict:
  indices[key]=x
  x+=1 
```

```
indices
```

index mapping to words


```
index_to_words = {}
x=0
for key in my_dict:
  index_to_words[x]=key
  x+=1 
```

```
index_to_words
```

```
for key in my_dict:
  flag = 0
  for word in only_words:
    if flag ==1:
      Bigram_matix[indices[key]][indices[word]]= Bigram_matix[indices[key]][indices[word]] + 1
      flag=0
    if key==word:
      flag=1
```

Functio to print frequency of occurence of next word based on the given word

```
def freq_of_occurence(word):
  print('Given word is',word)
  for i in indices:
    print('Frequency of occurence of', i,' is :', Bigram_matix[indices[word]][indices[i]])
  return Bigram_matix[indices[word]]

```
