#2) Room cleaner robot


Generating environment, i.e, Room


```
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
```

```
def make_room(rows, columns):
  room = np.zeros([rows,columns], dtype = 'int64')
  num=10
  for i in range(rows):
    for j in range(columns):
      if np.random.randint(1,1000)<300 and num>0:
        room[i][j] = np.random.randint(1,1000)
        num = num-1;
  return room
```

Creating agent function

```
def agent(observation):
  return np.random.randint(1,6)

# Note:
# Here the action being returned/ performed by the agent is in numbers between 1 to 5 
# 1 stands for "up"
# 2 stands for "down"
# 3 stands for "left"
# 4 stands for "right"
# 5 stands for "pick-dirt"
```

Creating Environment function

```
def environment(action, observation):
  reward=0
  if action==1:
    if observation[0]-1>=0 :
      observation[0]=observation[0]-1
    else:
      reward=reward-10
  elif action==2:
    if observation[0]+1<rows :
      observation[0]=observation[0]+1
    else:
      reward=reward-10
  elif action==3:
    if observation[1]-1>=0 :
      observation[1]=observation[0]-1
    else:
      reward=reward-10
  elif action==4:
    if observation[1]+1<columns :
      observation[1]=observation[1]+1
    else:
      reward=reward-10
  elif action==5:
    if room[observation[0]][observation[1]]!=0 :
      reward=reward+ room[observation[0]][observation[1]]
      room[observation[0]][observation[1]]=0
    else:
      reward=reward-1

  return observation, reward 
```
